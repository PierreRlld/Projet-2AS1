Nous allons donc "aligner" à la main les noms des mouvements.

```python
import pandas as pd
import wikipedia
from tqdm import tqdm
```

```python
# C'est le fichier data.csv
auteurs = pd.read_csv('622_authors.csv',index_col=[0])
```

```python
def page(s):
    try:
        page  = wikipedia.page(s)
        return page.content
    except:
        return ''
```

```python
#page('Lord Byron')
```

```python
auteurs['page']=auteurs['Nom'].apply(page)
```

```python
import spacy
import nltk
import numpy as np
nlp = spacy.load("fr_core_news_sm")
from nltk.corpus import stopwords
nltk.download('stopwords')
stop=stopwords.words('french')

def clean_txt(s):
    if s=='':
        return s
    doc = nlp(s)
    L=[]
    for token in doc:
        s_t = str(token)
        alphanum=True
        if not s_t in stop:
            for char in s_t:
                alphanum = char.isalnum() and alphanum
            if alphanum:            
                L.append(token.lemma_)
    L=np.array(L)
    return ' '.join(L)
        
    

```

```python
auteurs['page_lem']=auteurs['page'].apply(clean_txt)
```
